```python
gs0 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2010.csv')
gs1 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2011.csv')
gs2 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2012.csv')
gs3 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2013.csv')
gs4 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2014.csv')
gs5 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2015.csv')
gs6 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2016.csv')
gs7 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2017.csv')
gs8 = pd.read_csv('/kaggle/input/gen-crime-stat/gender_state_2018.csv')
```

```python
gs0.rename(columns={'ddl_case_id': '2010'}, inplace=True)
gs1.rename(columns={'ddl_case_id': '2011'}, inplace=True)
gs2.rename(columns={'ddl_case_id': '2012'}, inplace=True)
gs3.rename(columns={'ddl_case_id': '2013'}, inplace=True)
gs4.rename(columns={'ddl_case_id': '2014'}, inplace=True)
gs5.rename(columns={'ddl_case_id': '2015'}, inplace=True)
gs6.rename(columns={'ddl_case_id': '2016'}, inplace=True)
gs7.rename(columns={'ddl_case_id': '2017'}, inplace=True)
gs8.rename(columns={'ddl_case_id': '2018'}, inplace=True)
```

```python
m1 = gs0.merge(gs1, left_on="st_nm", right_on="st_nm", how="left")
m2 = m1.merge(gs2, left_on="st_nm", right_on="st_nm", how="left")
m3 = m2.merge(gs3, left_on="st_nm", right_on="st_nm", how="left")
m4 = m3.merge(gs4, left_on="st_nm", right_on="st_nm", how="left")
m5 = m4.merge(gs5, left_on="st_nm", right_on="st_nm", how="left")
m6 = m5.merge(gs6, left_on="st_nm", right_on="st_nm", how="left")
m7 = m6.merge(gs7, left_on="st_nm", right_on="st_nm", how="left")
m8 = m7.merge(gs8, left_on="st_nm", right_on="st_nm", how="left")
m8.set_index('st_nm', inplace=True)
print(m8)
```

```python
m8 = m8[(m8['2010'] > 0) | (m8['2011'] > 0) | (m8['2012'] > 0) | (m8['2013'] > 0) | (m8['2014'] > 0) | (m8['2015'] > 0) | (m8['2016'] > 0) | (m8['2017'] > 0) | (m8['2018'] > 0)]
print(m8)
```

```python
m20 = m8.iloc[:, 0:].T
print(m20)
```

```python
ax = m20.plot(xlabel='Years', ylabel='Cases',figsize=(30,30), marker='.')
ax.set_xlabel('Years', fontsize = 45)
ax.set_ylabel('Cases', fontsize = 45)
ax.legend(title='Sample', bbox_to_anchor=(1, 1.02), loc='upper left', prop={'size':26})
plt.rcParams.update({'font.size': 35})
plt.pyplot.savefig('gender_state_cumm_trend.png')
```
