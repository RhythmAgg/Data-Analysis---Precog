```python
fpet1 = pd.read_csv('/kaggle/input/gen-crime/fpet_2018.csv')
fpet2 = pd.read_csv('/kaggle/input/gen-crime/fpet_2017.csv')
fpet3 = pd.read_csv('/kaggle/input/gen-crime/fpet_2016.csv')
fpet4 = pd.read_csv('/kaggle/input/gen-crime/fpet_2015.csv')
fpet5 = pd.read_csv('/kaggle/input/gen-crime/fpet_2014.csv')
fpet6 = pd.read_csv('/kaggle/input/gen-crime/fpet_2013.csv')
```

```python
fp1 = fpet1.merge(fpet2, left_on="female_petitioner", right_on="female_petitioner", how="left", left_index=False, right_index=False)
fp2 = fp1.merge(fpet3, left_on="female_petitioner", right_on="female_petitioner", how="left", left_index=False, right_index=False)
fp3 = fp2.merge(fpet4, left_on="female_petitioner", right_on="female_petitioner", how="left", left_index=False, right_index=False)
fp4 = fp3.merge(fpet5, left_on="female_petitioner", right_on="female_petitioner", how="left", left_index=False, right_index=False)
fp5 = fp4.merge(fpet6, left_on="female_petitioner", right_on="female_petitioner", how="left", left_index=False, right_index=False)
```

```python
print(fp5)
```

```python
fp5.drop(fp5.columns[[0,3,5,7,9,11]], axis=1, inplace=True)
```

```python
print(fp5)
```

```python
print(fp5)
```

```python
fp5_t = fp5.set_index('female_petitioner').transpose()
```

```python
print(fp5_t)
```

```python
ax = fp5_t.plot.bar(color=["SkyBlue","IndianRed"],rot = 0, title="Gender of Petitioner in Gender Crimes over Years")
ax.set_xlabel("Years")
ax.set_ylabel("Number of Petitioners")

plt.pyplot.savefig('fpet.png')
```

```python
fdef1 = pd.read_csv('/kaggle/input/gen-crime/fdef_2018.csv')
fdef2 = pd.read_csv('/kaggle/input/gen-crime/fdef_2017.csv')
fdef3 = pd.read_csv('/kaggle/input/gen-crime/fdef_2016.csv')
fdef4 = pd.read_csv('/kaggle/input/gen-crime/fdef_2015.csv')
fdef5 = pd.read_csv('/kaggle/input/gen-crime/fdef_2014.csv')
fdef6 = pd.read_csv('/kaggle/input/gen-crime/fdef_2013.csv')
```

```python
fd1 = fdef1.merge(fdef2, left_on="female_defendant", right_on="female_defendant", how="left", left_index=False, right_index=False)
fd2 = fd1.merge(fdef3, left_on="female_defendant", right_on="female_defendant", how="left", left_index=False, right_index=False)
fd3 = fd2.merge(fdef4, left_on="female_defendant", right_on="female_defendant", how="left", left_index=False, right_index=False)
fd4 = fd3.merge(fdef5, left_on="female_defendant", right_on="female_defendant", how="left", left_index=False, right_index=False)
fd5 = fd4.merge(fdef6, left_on="female_defendant", right_on="female_defendant", how="left", left_index=False, right_index=False)
```

```python
print(fd5)
```

```python
fd5.drop(fd5.columns[[0,3,5,7,9,11]], axis=1, inplace=True)
```
